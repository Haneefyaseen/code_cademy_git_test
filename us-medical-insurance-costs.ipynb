# U.S. Medical Insurance Costs

```python
import csv

ages = []
_sex = []
bmi = []
children = []
uncategorized_smoker = []
region = []
charges = []

#Opening the csv file as dictionary and appending the values to separate lists
with open('insurance.csv') as insurance_file:
    insurance_file = csv.DictReader(insurance_file)
    for row in insurance_file:
        ages.append(row['age'])
        _sex.append(row['sex'])
        bmi.append(row['bmi'])
        children.append(row['children'])
        uncategorized_smoker.append(row['smoker'])
        region.append(row['region'])
        charges.append(row['charges'])
        
#Finding out the average of ages of patients
avg_age, sum_age = 0, 0
for i in ages:
    sum_age += float(i)
avg_age = sum_age/len(ages)
print("The average age of patients is {}\n".format(avg_age))

#Converting the yes / no values to 1 / 0
smoker = []
for i in uncategorized_smoker:
    if i == 'yes':
        smoker.append(1)
    else:
        smoker.append(0)

sex = []
for i in _sex:
    if i == 'male':
        sex.append(1)
    else:
        sex.append(0)
        
#Converting the data in the csv file into a dictionary
class Patient:
    def convert_to_dict(self, file):
        patient_records = []
        with open(file) as df:
            df = csv.DictReader(df)
            for row in df:
                patient_records.append(row)
        return patient_records
    
    
    #Categorizing binary values in the dictionary
    def categorized_data(self, upd_rec):
        
        #Converting the smokers to categorical data
        for i in upd_rec:
            if i['smoker'] == 'yes':
                i['smoker'] = 1
            else:
                i['smoker'] = 0
                
        #Converting gender to categorical data
        for i in upd_rec:
            if i['sex'] == 'male':
                i['sex'] = 1
            else:
                i['sex'] = 0
                
        return upd_rec
    
    #Calculating estimated insurance cost using formula
    def estimated_insurance_costs(self, record):
        insurance_cost = []
        for i in record:
            estimated_insurance_cost = round((((int(i['age']) * 250)  - (128 * int(i['sex'])) + (370 * float(i['bmi'])) + (425 * int(i['children'])) + (24000 * int(i['smoker'])) - 12500)), 2)
            insurance_cost.append(estimated_insurance_cost) 
        return insurance_cost
    
        #estimated_cost=250∗age−128∗sex+370∗bmi+425∗num_of_children+24000∗smoker−12500
    
#Creating object for the class Patient
patient = Patient()

patient_records = patient.convert_to_dict('insurance.csv')
#print("The uncategorized record of patients:\n{}\n\n\n".format(patient_records))

updated_patient_records = patient.categorized_data(patient_records)
#print("The categorized record of patients:\n{}".format(updated_patient_records))

estimated_insurance_cost = patient.estimated_insurance_costs(updated_patient_records)
print(estimated_insurance_cost)
```
