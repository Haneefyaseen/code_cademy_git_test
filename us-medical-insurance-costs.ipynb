# U.S. Medical Insurance Costs

```python
import csv
import pandas as pd

ages = []
_sex = []
bmi = []
children = []
uncategorized_smoker = []
region = []
charges = []

#Opening the csv file as dictionary and appending the values to separate lists
with open('insurance.csv') as insurance_file:
    insurance_file = csv.DictReader(insurance_file)
    for row in insurance_file:
        ages.append(row['age'])
        _sex.append(row['sex'])
        bmi.append(row['bmi'])
        children.append(row['children'])
        uncategorized_smoker.append(row['smoker'])
        region.append(row['region'])
        charges.append(row['charges'])
        
#Converting the yes / no values to 1 / 0
smoker = []
for i in uncategorized_smoker:
    if i == 'yes':
        smoker.append(1)
    else:
        smoker.append(0)

sex = []
for i in _sex:
    if i == 'male':
        sex.append(1)
    else:
        sex.append(0)

df = pd.read_csv('insurance.csv')
```
